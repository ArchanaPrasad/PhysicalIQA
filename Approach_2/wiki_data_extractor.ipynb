```python
premises1 = [ques + ' ' + sol for ques,sol in zip(premises,sol1)]
premises2 = [ques + ' ' + sol for ques,sol in zip(premises,sol2)]
```

```python
data = []
# premises = ['sadjkhasfiuahf;asfh']
for query in tqdm(premises1):
#     query = cleaned[i]
    temp = []
    summary = "None"
    result = "None"
    try:
        results = wikipedia.search(query)
    #     print(query + ': ')
    #     print(results)
        index = 0
        flag = True
        while flag and index<len(results):
            try:
                result = results[index]
    #             print(result)
                wiki_page = wikipedia.page(result, auto_suggest=False)
    #             print('here')
                summary = wiki_page.summary
                summary = summary.replace('\n','')
    #             print(query)
    #             print(summary)
    #             print('\n')
                flag=False
            except:
                index+=1
    #             print('error in wiki try block')
    #             print("Error")
    except:
        pass
    
    temp.append(query)
    temp.append(result)
    temp.append(summary)
    
    data.append(temp)
    


# using try blocks to catch exceptions, and then using wikipedia.suggest in the exception handling

# also try for DV, other spell checkers and abbreviations checker library for python
```
